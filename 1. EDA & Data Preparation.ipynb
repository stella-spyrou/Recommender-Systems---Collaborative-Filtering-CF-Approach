# Data Preparation 

### And Feature Engineering will follow

```python
#Map MCC to Master Categories

def mcc_mapping(row):
    if row['mcc_code'] >= 0 and row['mcc_code'] < 1500:
        return 'Agricultural Services'
    elif row['mcc_code'] >= 1500 and row['mcc_code'] < 3000:
        return 'Contracted Services'
    elif row['mcc_code'] >= 3000 and row['mcc_code'] < 3300:
        return 'Airlines'
    elif row['mcc_code'] >= 3300 and row['mcc_code'] < 3500:
        return 'Car Rental'
    elif row['mcc_code'] >= 3500 and row['mcc_code'] < 4000:
        return 'Lodging'
    elif row['mcc_code'] >= 4000 and row['mcc_code'] < 4800:
        return 'Transportation Services'
    elif row['mcc_code'] >= 4800 and row['mcc_code'] < 5000:
        return 'Utility Services'
    elif row['mcc_code'] >= 5000 and row['mcc_code'] < 5600:
        return 'Retail Outlet Services'
    elif row['mcc_code'] >= 5600 and row['mcc_code'] < 5700:
        return 'Clothing Stores'
    elif row['mcc_code'] >= 5700 and row['mcc_code'] < 7300:
        return 'Miscellaneous Stores'
    elif row['mcc_code'] >= 7300 and row['mcc_code'] < 8000:
        return 'Business Services'
    elif row['mcc_code'] >= 8000 and row['mcc_code'] < 9000:
        return 'Professional Services and Membership Organizations'
    elif row['mcc_code'] >= 9000 and row['mcc_code'] <= 9999:
        return 'Government Services'
    else:
        return 'Other'

def mcc_to_categories(df):       #I rename this in the next cell cmd
  df = df.copy()
  df['category'] = df.apply(mcc_mapping, axis=1)
  return df
```

```python
#Rename table - including 'category' column (MCC category aggregation)
transactions_cat = mcc_to_categories(transactions_pd)
```

```python
#Display Transactions table along with the assigned MCC Master Category (Aggregation)

#display(mcc_to_categories(transactions_pd))
display(transactions_cat.head())
```

```python
#Looking the type of the new variable 'category' - which initially is 'object' 

transactions_cat.dtypes
```

```python
#Transform the type of the variable 'category' from __'object'__ to __'category'__

transactions_cat.category = pd.Categorical(transactions_cat.category)
```

```python
#Creating new variable - 'categoryId' (useful for Modelling - RecSys)
#This way assigns on the 9 master categories indeces ( i.e. from 0 to 8 )

transactions_cat['categoryId'] = transactions_cat.category.cat.codes
```

```python
#Displaying the table (observing the new column categoryId)
transactions_cat.head()
```

# Feature Engineering

This section of Feature Engineering, takes place because we need to construct the proportion of spend. This will be done by finding first the ***sum of the tran_amount per customer***. Secondly, finding the ***sum of the tran_amount per category and per customer***. Later, we *divide* the sum of tran_amount per category & per customer with the sum of the tran_amount per customer. In this way we will create a new column that will be the **proportion of spend** which we will need in order to **feed the algorithm** (as a metric of customer preference) since in this dataset we don't have ratings or some other similar metric but customer transactions.

```python
#create the sum of the transaction amount per customer

sum_amount_per_cust = transactions_cat.groupby(['customer_id']).agg({'tran_amount': 'sum'})
```

```python
#Reseting the index after grouping
sum_amount_per_cust_df = pd.DataFrame(sum_amount_per_cust.reset_index())

#renaming the tran_amount column to reflect the aggregation per customer
sum_amount_per_cust_df.rename(columns={'tran_amount': 'SumAmount_PerCustomer'}, inplace=True)
sum_amount_per_cust_df.head()
```

```python
#Grouping by customer (9,000 distinct) and master category (9 distinct) 
#and finding the sum of transaction amount per category and per customer

sum_amount_per_cat_per_cust = transactions_cat.groupby(['customer_id','category', 'categoryId']).agg({'tran_amount': 'sum'})
```

```python
#renaming
sum_amount_per_cat_per_cust.rename(columns={'tran_amount': 'SumAmount_PerCat_PerCust'}, inplace=True)
```

```python
#Result after grouping - seems I have only 1 column (transformation below) - tran_amount is the SUM 
#After grouping - we can see the SUM of the 'tran_amount' variable per 'customer_id' and per 'category'

sum_amount_per_cat_per_cust.head(5)
```

```python
#Printing the table: Distinct [customer_id & category], sum(tran_amount)

#NOTE: in the previous command/output created a table with only one column since data are grouped using the 3 variables
#the below action will transform the table - creating a table with 4 columns

rec_sys_df = pd.DataFrame(sum_amount_per_cat_per_cust.reset_index())
rec_sys_df.head()
```

```python
rec_sys_df.shape
```

```python
#Merging the tables ([rec_sys_df], [sum_amount_per_cust_df]) on customer_id 
#in order to find the proportion spend per customer and per category

rec_sys_merge=rec_sys_df.merge(sum_amount_per_cust_df, left_on='customer_id', right_on='customer_id', how='inner')
rec_sys_merge.head()
```

```python
rec_sys_merge.shape
```

```python
#part of the Feature Engineering - constructing the proportion of spend (with the division of the two columns created)

rec_sys_merge['proportion_spent'] = rec_sys_merge['SumAmount_PerCat_PerCust']/rec_sys_merge['SumAmount_PerCustomer']
rec_sys_merge.head(7)
```

```python
#Filtering for a specific customer based on the previous output table 
#we see all transactions and total amount spend across the categories associated with the 'x' custmer

is_mycust = rec_sys_merge['customer_id']== '227714a3-e0a9-4260-8d67-fc22665b4992'    #'00015fc4-6fdc-4120-be42-a49e8bbe8cce', '0016ee8d-e2d6-4dd9-9559-3be13e9f53b5'
print(is_mycust.head())
rec_sys_merge_is_mycust = rec_sys_merge[is_mycust]
print(rec_sys_merge_is_mycust.shape)
rec_sys_merge_is_mycust
```

```python
#  Saving to csv my new created table - processed and prepared for Data Modelling

rec_sys_merge.to_csv('/dbfs/mnt/data/sandbox/proportion_spent_rec_sys.csv', index=False)
```
